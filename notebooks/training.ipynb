# Dynamic CSV Visualization Notebook

This notebook allows you to dynamically load a CSV file and generate visualizations for its columns. You can explore your data interactively and create various plots based on the data types.

## 1. Import Required Libraries
We will use pandas for data handling and seaborn/matplotlib for visualization. ipywidgets will be used for interactivity.

```python
!pip install pandas matplotlib seaborn ipywidgets
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display, clear_output
```

## 2. Load CSV File Dynamically
You can specify the path to your CSV file below. The default is set to 'cement_steel_sand_only.csv'.

```python
csv_path = widgets.Text(
    value='cement_steel_sand_only.csv',
    description='CSV Path:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)
load_button = widgets.Button(description='Load CSV')
output = widgets.Output()

def load_csv(b):
    with output:
        clear_output()
        try:
            df = pd.read_csv(csv_path.value)
            print('CSV loaded successfully!')
            display(df.head())
            globals()['df'] = df  # Store in global scope for later cells
        except Exception as e:
            print(f'Error loading CSV: {e}')

load_button.on_click(load_csv)
display(csv_path, load_button, output)
```

## 3. Display Data Overview
Once the CSV is loaded, the following cell will show the first few rows, column names, and basic statistics.

```python
try:
    df
except NameError:
    print('Please load the CSV file first using the cell above.')
else:
    display(df.head())
    print('\nColumn Names:')
    print(df.columns.tolist())
    print('\nData Types:')
    print(df.dtypes)
    print('\nBasic Statistics:')
    display(df.describe(include='all'))
```

## 4. Generate Visualizations Based on Data
This cell will automatically generate basic plots for each column based on its data type.

```python
try:
    df
except NameError:
    print('Please load the CSV file first using the cell above.')
else:
    for col in df.columns:
        plt.figure(figsize=(6, 3))
        if pd.api.types.is_numeric_dtype(df[col]):
            sns.histplot(df[col].dropna(), kde=True)
            plt.title(f'Histogram of {col}')
        elif pd.api.types.is_categorical_dtype(df[col]) or df[col].dtype == object:
            vc = df[col].value_counts().head(20)
            sns.barplot(x=vc.index, y=vc.values)
            plt.title(f'Bar Plot of {col} (Top 20)')
            plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

## 5. Interactive Visualization Selection
Use the widgets below to select columns and plot types for custom visualizations.

```python
try:
    df
except NameError:
    print('Please load the CSV file first using the cell above.')
else:
    col_options = df.columns.tolist()
    plot_types = ['Histogram', 'Bar Plot', 'Line Plot', 'Box Plot']
    col_dropdown = widgets.Dropdown(options=col_options, description='Column:')
    plot_dropdown = widgets.Dropdown(options=plot_types, description='Plot Type:')
    out = widgets.Output()

    def plot_interactive(change=None):
        with out:
            clear_output()
            col = col_dropdown.value
            plot_type = plot_dropdown.value
            plt.figure(figsize=(6, 3))
            if plot_type == 'Histogram' and pd.api.types.is_numeric_dtype(df[col]):
                sns.histplot(df[col].dropna(), kde=True)
                plt.title(f'Histogram of {col}')
            elif plot_type == 'Bar Plot':
                vc = df[col].value_counts().head(20)
                sns.barplot(x=vc.index, y=vc.values)
                plt.title(f'Bar Plot of {col} (Top 20)')
                plt.xticks(rotation=45)
            elif plot_type == 'Line Plot' and pd.api.types.is_numeric_dtype(df[col]):
                plt.plot(df[col].dropna().values)
                plt.title(f'Line Plot of {col}')
            elif plot_type == 'Box Plot' and pd.api.types.is_numeric_dtype(df[col]):
                sns.boxplot(x=df[col].dropna())
                plt.title(f'Box Plot of {col}')
            else:
                print('Selected plot type is not compatible with this column.')
            plt.tight_layout()
            plt.show()

    col_dropdown.observe(plot_interactive, names='value')
    plot_dropdown.observe(plot_interactive, names='value')
    display(widgets.HBox([col_dropdown, plot_dropdown]))
    display(out)
    plot_interactive()
```

```python
# ================================
# STEP 0: Safety check
# ================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

if 'df' not in globals():
    raise ValueError("Please load the CSV into variable `df` first.")

# ================================
# STEP 1: Identify index columns
# ================================
index_cols = [c for c in df.columns if c.startswith("indx")]

# ================================
# STEP 2: Wide → Long reshape
# ================================
df_long = (
    df
    .melt(
        id_vars=["comm_name", "comm_code", "comm_wt"],
        value_vars=index_cols,
        var_name="month",
        value_name="price_index"
    )
)

# Convert month string like indx042012 → datetime
df_long["month"] = (
    df_long["month"]
    .str.replace("indx", "", regex=False)
)
df_long["date"] = pd.to_datetime(df_long["month"], format="%m%Y")

df_long = df_long.sort_values(["comm_name", "date"]).reset_index(drop=True)

print("Reshaped data:")
display(df_long.head())

# ================================
# STEP 3: Choose construction-relevant commodities
# ================================
KEYWORDS = [
    "steel",
    "iron",
    "bars",
    "rods",
    "alloy",
    "metal"
]

df_long["is_construction"] = df_long["comm_name"].str.lower().apply(
    lambda x: any(k in x for k in KEYWORDS)
)

construction_df = df_long[df_long["is_construction"]].copy()

print("\nSelected construction-related commodities:")
display(construction_df["comm_name"].unique())

# ================================
# STEP 4: Create trend label (target)
# ================================
construction_df["pct_change"] = (
    construction_df
    .groupby("comm_name")["price_index"]
    .pct_change()
)

# Trend definition:
# +1 → increase, 0 → stable, -1 → decrease
THRESHOLD = 0.003  # 0.3%

construction_df["trend"] = np.where(
    construction_df["pct_change"] > THRESHOLD, 1,
    np.where(construction_df["pct_change"] < -THRESHOLD, -1, 0)
)

construction_df = construction_df.dropna().reset_index(drop=True)

# ================================
# STEP 5: Feature engineering
# ================================
construction_df["lag_1"] = (
    construction_df.groupby("comm_name")["price_index"].shift(1)
)
construction_df["lag_3_mean"] = (
    construction_df.groupby("comm_name")["price_index"]
    .rolling(3)
    .mean()
    .reset_index(level=0, drop=True)
)

construction_df = construction_df.dropna()

X = construction_df[["price_index", "lag_1", "lag_3_mean"]]
y = construction_df["trend"]

# ================================
# STEP 6: Train / Test split (time-based)
# ================================
split_date = construction_df["date"].quantile(0.8)

train_mask = construction_df["date"] <= split_date
test_mask  = construction_df["date"] > split_date

X_train, X_test = X[train_mask], X[test_mask]
y_train, y_test = y[train_mask], y[test_mask]

# ================================
# STEP 7: Train first working model
# ================================
model = LogisticRegression(
    multi_class="multinomial",
    max_iter=500
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("\nModel Performance:")
print(classification_report(y_test, y_pred))
import joblib

# Save trained model
joblib.dump(model, "trend_model.pkl")

# Save feature order (VERY IMPORTANT)
joblib.dump(X.columns.tolist(), "model_features.pkl")

print("✔ Model and feature schema saved successfully")

# ================================
# STEP 8: Visualization (one commodity)
# ================================
sample_comm = construction_df["comm_name"].iloc[0]
plot_df = construction_df[construction_df["comm_name"] == sample_comm]

plt.figure(figsize=(10, 4))
plt.plot(plot_df["date"], plot_df["price_index"], label="Price Index")
plt.title(f"Price Index Trend: {sample_comm}")
plt.xlabel("Date")
plt.ylabel("Index Value")
plt.legend()
plt.grid(True)
plt.show()

print("\nMODEL READY ✔")
print("Output meaning:")
print("  1  → Price likely to increase")
print("  0  → Price likely stable")
print(" -1  → Price likely decrease")
```

```python
# ======================================================
# ENSEMBLE MODEL: Logistic + XGBoost + LightGBM (GPU)
# ======================================================

import numpy as np
import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# ------------------------------
# SAFETY CHECK
# ------------------------------
required_cols = {"price_index", "lag_1", "lag_3_mean", "trend", "date"}
if not required_cols.issubset(construction_df.columns):
    raise ValueError("Required columns missing. Run feature engineering first.")

X = construction_df[["price_index", "lag_1", "lag_3_mean"]]
y = construction_df["trend"]

# ------------------------------
# TIME-AWARE SPLIT
# ------------------------------
split_date = construction_df["date"].quantile(0.8)

train_mask = construction_df["date"] <= split_date
test_mask  = construction_df["date"] > split_date

X_train, X_test = X[train_mask], X[test_mask]
y_train, y_test = y[train_mask], y[test_mask]

# ------------------------------
# MODEL 1: Logistic Regression
# ------------------------------
log_reg = Pipeline([
    ("scaler", StandardScaler()),
    ("lr", LogisticRegression(
        max_iter=500,
        n_jobs=-1
    ))
])

# ------------------------------
# MODEL 2: XGBoost (GPU)
# ------------------------------
xgb = XGBClassifier(
    objective="multi:softprob",
    num_class=3,
    n_estimators=300,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    tree_method="hist",      # GPU
    predictor="gpu_predictor",   # GPU
    eval_metric="mlogloss",
    verbosity=0
)

# ------------------------------
# MODEL 3: LightGBM (GPU)
# ------------------------------
lgbm = LGBMClassifier(
    objective="multiclass",
    num_class=3,
    n_estimators=300,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    device="gpu",        # GPU
    gpu_use_dp=False
)

# Remap trend labels: -1 → 0, 0 → 1, 1 → 2
y_train_xgb = y_train.map({-1: 0, 0: 1, 1: 2})
y_test_xgb = y_test.map({-1: 0, 0: 1, 1: 2})

# Train models
print("Training Logistic Regression...")
log_reg.fit(X_train, y_train)

print("Training XGBoost (GPU)...")
xgb.fit(X_train, y_train_xgb)

print("Training LightGBM (GPU)...")
lgbm.fit(X_train, y_train_xgb)

# Predict probabilities
proba_lr  = log_reg.predict_proba(X_test)
proba_xgb = xgb.predict_proba(X_test)
proba_lgb = lgbm.predict_proba(X_test)

# Ensemble prediction
ensemble_proba = (
    0.9 * proba_lr +
    0.5 * proba_xgb +
    0.5 * proba_lgb
)
ensemble_pred = np.argmax(ensemble_proba, axis=1)

# Map back to original labels
reverse_map = {0: -1, 1: 0, 2: 1}
ensemble_pred_labels = np.array([reverse_map[i] for i in ensemble_pred])

# Evaluation
print("\nENSEMBLE MODEL PERFORMANCE:")
print(classification_report(y_test, ensemble_pred_labels))

# ------------------------------
# CONFIDENCE SCORE (PER SAMPLE)
# ------------------------------
confidence_scores = ensemble_proba.max(axis=1)

print("\nSample Predictions with Confidence:")
for i in range(5):
    print({
        "prediction": int(ensemble_pred_labels[i]),
        "confidence": round(confidence_scores[i], 3)
    })

print("\nENSEMBLE MODEL READY ✔")
print("Trend labels:")
print("  1  → Price likely to increase")
print("  0  → Price likely stable")
print(" -1  → Price likely decrease")
```
